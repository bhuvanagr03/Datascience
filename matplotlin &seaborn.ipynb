```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('/content/Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('Satscore')
plt.ylabel('Frequency')
plt.title('Distrubution of satscore')
plt.show()
```

```python
sns.histplot(df1['SAT'] , color= 'green')
plt.show()
```

```python
plt.plot(df1['SAT'] , df1['Accept'], 'bs')
plt.xlabel('SATscore')
plt.ylabel('Acceptance Rate')
plt.title('Acceptance Rate VS SATscore')
plt.show()
```

```python
plt.scatter(df1['SAT'], df1['Top10'])
plt.xlabel('SAT Score')
plt.ylabel('Top10 Percentage')
plt.title('SAT Score vs. Top10 Percentage')
plt.show()
```

```python
sns.pairplot(df1)
```

```python
plt.boxplot(df1['SAT'])
plt.xlabel('SAT Score')
plt.ylabel('Distribution')
plt.title('Box Plot of SAT Scores')
plt.show()
```
